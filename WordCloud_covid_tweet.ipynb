```python
data_conf_1 = pd.read_csv("../input/datacov2/tweets_conf_1.csv")
data_conf_1.head()
```

```python
data_conf_2 = pd.read_csv("/kaggle/input/datacov2/tweets_conf_2.csv")
data_conf_2.head()
```

```python
# Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of word
text_conf_1 = data_conf_1['text'].astype(str).tolist()

# Create show_wordcloud method:
def show_wordcloud(data, title):
    text_conf_1 = ' '.join(data_conf_1['text'].astype(str).tolist())
    #print (type(text))
           
# Create the wordcloud object
    fig_wordcloud = WordCloud(background_color='lightgrey',
                            colormap='viridis', width=800, height=600).generate(text_conf_1)
    
# Display the generated image:
    plt.figure(figsize=(10,7), frameon=True)
    plt.imshow(fig_wordcloud)
    plt.axis('off')
    plt.title(title, fontsize=20)
    plt.show()
          
show_wordcloud(data_conf_1,"Nuage de mot du 1er confinement")

# Save the image in the img folder:
#wordcloud.to_file("img/first_review.png")
```

```python
# WordCloud for 2nd confinement
text_conf_2 = data_conf_2['text'].astype(str).tolist()

def show_wordcloud_2(data, title):
    text_conf_2 = ' '.join(data_conf_2['text'].astype(str).tolist()) 
    
    fig_wordcloud_2 = WordCloud(background_color='lightgrey',
                            colormap='viridis', width=800, height=600).generate(text_conf_2)
    
    plt.figure(figsize=(10,7), frameon=True)
    plt.imshow(fig_wordcloud_2)
    plt.axis('off')
    plt.title(title, fontsize=20)
    plt.show()  
    
show_wordcloud_2(data_conf_2,"Nuage de mot du 2er confinement")
```
