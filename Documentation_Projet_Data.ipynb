# Documentation Projet Data Science

```python
import pandas as pd
import numpy as np
import glob
from pandas import DataFrame
import matplotlib.pyplot as plt
import seaborn as sns

path = r'F:\Ynov\Master 2\TP_COVID_YNOV-master\covid_departements_datasets' # use your path
 
all_files = glob.glob(path + "/*.csv")
 
li = []
 
for filename in all_files:
    df = pd.read_csv(filename)
    df.columns = ['departement', 'date', 'hospitalises', 'reanimations', 'nvx_hospitalises', 'nvlles_reanimations', 'gueris', 'deces']
 
    print(df)
    li.append(df)
 
Departements = pd.concat(li, ignore_index=True)
```

```python
sns.set(style="ticks", color_codes=True, font_scale=1.5)
color = sns.color_palette()
sns.set_style('darkgrid')

sns.relplot(x="date", y="hospitalises",
            col="departement", col_wrap=5,
            kind="line", data=Departements);
```
